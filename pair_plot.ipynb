# ペアプロットは、共分散の傾向を可視化している

```python
import pandas as pd
import seaborn as sns
pd.set_option('display.max_rows', None)
iris = sns.load_dataset('iris')
iris
```

```python

#plt.rcParams["figure.figsize"] = (10,10)
sns.pairplot(iris)

```

```python


temp_cov = iris.iloc[:,:4].cov()

cov=temp_cov.values

sns.heatmap(temp_cov, annot=True,fmt=".3f", square=True, annot_kws={"fontsize": 16, "fontweight": "bold", "fontfamily": "serif"})
```
